Generar un DataFrame con los datos del fichero

```python
import pandas as pd
df = pd.read_csv('titanic (1) (1).csv')
```

2. Mostrar por pantalla las dimensiones del DataFrame, el número de datos que contiene, los nombres de sus columnas y filas, los tipos de datos de las columnas, las 10 primeras filas y las 10 últimas filas.

```python
print("Dimensiones del Dataframe:", df.shape) #n de filas y columnas
print("Numero de datos:", df.size) # n de elementos
print("Nombres de las comunas:", df.columns.tolist())#lista de nombre de las columnas
print("Nombre de las filas (primeras 10)", df.index[:10]) #nombre primeras diez filas
print("Tipos de datos:", df.dtypes)
print("Primeras diez filas:", df. head(10))
print("Ultimas diez filas:", df.tail(10))

```

# 3. Setear el índice con la columna PassengerId

```python
df.columns = df.columns.str.strip()
df.set_index('PassengerId', inplace=True)
```

```python
print("Datos del pasajero con identificador 148:")
print(df.loc[148])
```

Edad promedio de los pasajeros

```python
print("Edad promedio de los pasajeros:", df['Age'].mean())
```

6. Pasajeros por clase

```python
print("Pasajeros por clase:")
print(df['Pclass'].value_counts())
```

Tarifa mas alta

```python
print("Tarifa mas alta pagada", df['Fare'].max())
```

8 Pasajeros embarcados por puerto

```python
print("Pasajeros por puerto:")
print(df['Embarked'].value_counts())
```

9. Nombre de las personas en primera clase en order alfabetico

```python
print("Nombres de las personas en primera clase ordenados alfabéticamente:")
print(df[df['Pclass'] == 1]['Name'].sort_values())
```


 10. Porcentaje de personas que sobrevivieron y murieron

```python
sobrevivieron = df['Survived'].value_counts(normalize=True) * 100
print("Porcentaje de personas que sobrevivieron y murieron:")
print(sobrevivieron)
```

 11. Añadir columna booleana para ver si el pasajero era menor de edad

```python
df['Menor_de_edad'] = df['Age'] < 18
print(df[['Age', 'Menor_de_edad']].head())
```

 12. Mostrar filas pares del DataFrame

```python
print("Filas pares del DataFrame:")
print(df.iloc[::2])
```

 13. Porcentaje de personas que sobrevivieron en cada clase

```python
print("Porcentaje de personas que sobrevivieron en cada clase:")
print(df.groupby('Pclass')['Survived'].mean() * 100)
```


# 14. Edad media de las mujeres en cada clase

```python
print("Edad media de las mujeres por clase:")
print(df[df['Sex'] == 'female'].groupby('Pclass')['Age'].mean())
```

15. Porcentaje de menores y mayores que sobrevivieron en cada clase

```python
print("Porcentaje de menores y mayores que sobrevivieron en cada clase:")
print(df.groupby(['Pclass', 'Menor_de_edad'])['Survived'].mean() * 100)
```


# 16. Eliminar pasajeros con edad desconocida

```python
df.dropna(subset=['Age'], inplace=True)
print("DataFrame después de eliminar pasajeros con edad desconocida:")
print(df.info())
```
